```python
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
x_train = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/IBM-Project-321-1658278283-main/Data Collection/Train_Set',
    target_size = (64, 64),
    batch_size = 300,
    class_mode = 'categorical',
    color_mode = 'grayscale'
)
```

```python
x_test = test_datagen.flow_from_directory(
    '/content/drive/MyDrive/IBM-Project-321-1658278283-main/Data Collection/Test_set',
    target_size = (64, 64),
    batch_size = 300,
    class_mode = 'categorical',
    color_mode = 'grayscale'
)
```
